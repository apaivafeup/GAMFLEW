```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.express as px
from plotly.offline import init_notebook_mode, iplot, plot
```

```python
q2 = pd.read_csv('q2.csv')
```

```python
column_names = ['date', 'code']
for i in range(1, 44):
    column_names.append(str(i))
```

```python
q2.columns = column_names
```

```python
strongly_agree = []
agree = []
neutral = []
disagree = []
strongly_disagree = []

for i in range(1, 43):
    # check if there is a value in the column
    if 'Strongly agree' in q2[str(i)].value_counts().index:
        strongly_agree.append(q2[str(i)].value_counts()['Strongly agree'])
    else:
        strongly_agree.append(0)

    if 'Agree' in q2[str(i)].value_counts().index:
        agree.append(q2[str(i)].value_counts()['Agree'])
    else:
        agree.append(0)

    if 'Neutral' in q2[str(i)].value_counts().index:
        neutral.append(q2[str(i)].value_counts()['Neutral'])
    else:
        neutral.append(0)

    if 'Disagree' in q2[str(i)].value_counts().index:
        disagree.append(q2[str(i)].value_counts()['Disagree'])
    else:
        disagree.append(0)

    if 'Strongly disagree' in q2[str(i)].value_counts().index:
        strongly_disagree.append(q2[str(i)].value_counts()['Strongly disagree'])
    else:
        strongly_disagree.append(0)
```

```python
# 0 to 10 
strongly_agree_10 = strongly_agree[:10]
agree_10 = agree[:10]
neutral_10 = neutral[:10]
disagree_10 = disagree[:10]
strongly_disagree_10 = strongly_disagree[:10]

data = pd.DataFrame({
    "Strongly agree": strongly_agree_10,
    "Agree": agree_10,
    "Neutral": neutral_10,
    "Disagree": disagree_10,
    "Strongly disagree": strongly_disagree_10},
    index = [str(i) for i in range(1, 11)]
)

data.plot(kind="bar",figsize=(15, 8))

plt.title("Questionnaire 2 Answers (1-10)")

plt.xlabel("Questions")

plt.ylabel("Count")
```
